```python
df = pd.concat([df_suisyo,df_tyui_to_suisyo])
df
```

```python
old = list(df['原則として使用しない表記'])
new = list(df['推奨表記'])

print(old)
print(new)
```

```python
file_name ='チェック対象メール文.txt'

with open(file_name, encoding='utf-8') as f:
    check_text = f.read()
    print(check_text)
```

```python
# 修正対象のテキスト：check_text
# 表現置換リスト：old, new

for x, y in zip(old, new):
#     check_text = check_text.replace(x,  y)
    y =  '【'+ x + '】　→ 【'+ y + '】'
    check_text = check_text.replace(x, y)
print(check_text)
```

```python
document = Document() #書き込み 
document.add_paragraph(check_text) #保存 
document.save('訂正後テキスト_添付用.docx')
```
